```python
#drop columns that are useless
df.drop(['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
#converting overtime to yes/no to 1/0
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
multi_cat_cols = ['Department','JobRole','MaritalStatus','EducationField','BusinessTravel']
df = pd.get_dummies(df, columns=multi_cat_cols, drop_first=True)
```

```python
#Feature Engineering
# ---------------------------
# Tenure-related features
df['PromotionGap'] = df['YearsSinceLastPromotion'] / df['YearsAtCompany'].replace(0,1)
df['RoleStability'] = df['YearsInCurrentRole'] / df['YearsAtCompany'].replace(0,1)

# Salary-related feature
df['IncomePerYear'] = df['MonthlyIncome'] * 12 / df['TotalWorkingYears'].replace(0,1)

# Satisfaction feature
df['OverallSatisfaction'] = df[['JobSatisfaction','EnvironmentSatisfaction',
                                'RelationshipSatisfaction','WorkLifeBalance']].mean(axis=1)

# Experience-related features
df['JobHoppingRatio'] = df['NumCompaniesWorked'] / df['TotalWorkingYears'].replace(0,1)
df['LoyaltyRatio'] = df['YearsAtCompany'] / df['TotalWorkingYears'].replace(0,1)
```

```python
df.to_csv("employee_attrition_cleaned.csv", index=False)
print("✅ Cleaned and feature-engineered CSV saved as 'employee_attrition_cleaned.csv'")
```
